It involves various statistical and visual techniques to gain a preliminary understanding of the data, identify outliers and missing values, and ultimately inform further analysis and decision-making. ## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# I have already dropped the 'company' Column so it is one of the wrangling code.

df_hotel['agent'].fillna(value = 0 , inplace = True) #as mentioned earlier filling null values of agent by '0'
print(df_hotel['agent'].isnull().sum()) #checking null values
````

````python
#filling null values in country column by 'other'
df_hotel['country'].fillna(value= 'other' , inplace = True)
df_hotel['country'].isnull().sum() #checking null values
````

````python
#filling null value in children column by '0' assuming not mentioned have no childrens during stays
df_hotel['children'].fillna(value=0 , inplace=True)
df_hotel['children'].isnull().sum() #checking null values
````

````python
#counting the rows wih duplicate values
print(df_hotel.duplicated().value_counts())
````

````python
#as we can see that the count of duplicates are in large number so it can mislead the results so it is better to drop them.
df_hotel.drop_duplicates(inplace=True)
df_hotel.reset_index(drop=True,inplace=True)
df_hotel['new_index'] = df_hotel.index
df_hotel.tail()
````

````python
# change the data type of agent and children from float to int
df_hotel[['children','agent']]= df_hotel[['children','agent']].astype('int64')
#changing the type of the data in the columns to datetime format
df_hotel['reservation_status_date'] = pd.to_datetime(df_hotel['reservation_status_date'],format = '%Y-%m-%d')
````

````python
#adding new columns the data frame as per requirement
df_hotel['total_days_stay']=df_hotel['stays_in_week_nights']+df_hotel['stays_in_weekend_nights']
df_hotel['total_person']=df_hotel['adults']+df_hotel[ 'babies']+df_hotel['children']

#total stay in night
df_hotel['total_stay_at_night'] = df_hotel['stays_in_week_nights']+df_hotel['stays_in_weekend_nights']


#we have created a cloumn for revenue using total stay*adr
df_hotel['revenue'] = df_hotel['total_stay_at_night']*df_hotel['adr']
````

````python
df_hotel.info()
````

### What all manipulations have you done and insights you found?

1. Data Type Conversion:
The data types of the 'agent' and 'children' columns were converted from float to integer using the astype('int64') method. This conversion is useful when these columns should be treated as discrete numerical values.

2. Date Conversion:
The 'reservation_status_date' column was converted to datetime format using the pd.to_datetime() method. This conversion allows for easier manipulation and analysis of date-related information.

3. New Columns:
Two new columns were added to the DataFrame:
'total_days_stay': This column represents the total number of days stayed by guests, calculated by summing the 'stays_in_week_nights' and 'stays_in_weekend_nights' columns. It provides valuable insights into the length of guests' stays, considering both weekday and weekend nights.
'total_person': This column represents the total number of persons associated with each booking, including adults, babies, and children. It offers insights into the group size and composition of guests for each reservation.

4. Total Stay in Nights:
An additional column named 'total_stay_at_night' was created by summing the 'stays_in_week_nights' and 'stays_in_weekend_nights' columns. This column indicates the total number of nights a guest stays at the hotel, combining both weekday and weekend stays.

5. Revenue Calculation:
The 'revenue' column was created by calculating the revenue generated from each booking. It was computed by multiplying the 'total_stay_at_night' with the 'adr' (Average Daily Rate). This newly added column provides insights into the revenue generated from each booking.


***Insights:***

The conversion of 'agent' and 'children' columns from float to integer implies that these columns contain discrete numerical values (IDs for the agent and number of children), making them suitable for various numerical operations.
The 'reservation_status_date' column's conversion to datetime format enables easier temporal analysis, such as finding patterns in reservation statuses over time.
The newly added columns 'total_days_stay' and 'total_person' offer valuable insights into the length of stays and the composition of guest groups, respectively, which can help hotel management tailor services accordingly.
The 'total_stay_at_night' column consolidates information about the total number of nights guests stayed, providing a clearer picture of their overall stay duration.
The 'revenue' column helps assess the revenue generated from each booking, facilitating revenue analysis and strategic decision-making related to pricing and revenue management.
